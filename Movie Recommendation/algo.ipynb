```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits,on='title')
```

```python
#genres
#id 
#keywords
# title
#overview
#popularity
# production companies 
#cast
#crew


movies = movies[['movie_id','title','overview','genres','keywords', 'cast', 'crew', 'production_companies',]]

```

```python
movies.dropna(inplace=True)
```

```python
import ast
ast.literal_eval
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
def convert2(obj):
    counter = 0;
    l=[]
    for i in ast.literal_eval(obj):
        if counter<3:
         l.append(i['name'])
         counter+=1
    return l
```

```python
movies['cast'] =  movies['cast'].apply(convert2)
```

```python
movies.head(1)
```

```python
movies["production_companies"][0]
```

```python
def director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job'] == "Director":
            l.append(i['name'])
            break
    return l
```

```python
movies['crew'] = movies['crew'].apply(director)
```

```python
movies.head(1)
```

```python
movies['production_companies'] =  movies['production_companies'].apply(convert2)
```

```python
movies.head(1)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(1)
```

```python
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(' ',"") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(' ',"") for i in x])
movies['production_companies'] = movies['production_companies'].apply(lambda x:[i.replace(' ',"") for i in x])
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(' ',"") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(' ',"") for i in x])
```

```python
movies.head(1)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast']+ movies['crew'] + movies['production_companies']
```

```python
movies.head(1)
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df.head(1)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head(1)
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df['tags'][0]
```

```python
new_df.head(1)
```

```python
new_df['title'] = new_df['title'].apply(lambda x:x.lower())
new_df.head(1)
```

```python
import nltk 
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)
```

```python
new_df['tags']  = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors  = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movies):
    
    movie_index = new_df[new_df['title']== movies.lower()].index[0]
    distance = similarity[movie_index]
    movies_list = sorted(list(enumerate(distance)),reverse=True,key = lambda x: x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title.title())
```

```python
recommend('Avatar')
```
