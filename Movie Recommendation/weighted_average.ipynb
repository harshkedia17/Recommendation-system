```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits  = pd.read_csv('tmdb_5000_credits.csv')
```

```python
new_df = movies.merge(credits,on='title')
```

```python
new_df = new_df.drop(columns = ['homepage','original_title','original_language','status','production_countries'])
```

Using Weighted average for each movie's Average Rating
![](http://trailerpark.weebly.com/uploads/8/8/5/5/8855465/7628808.png?371)

```python
v = new_df['vote_count']
R = new_df['vote_average']
C = new_df['vote_average'].mean()
m = new_df['vote_count'].quantile(0.70)
```

```python
new_df['weighted_average'] = ((R*v)+(C*m))/(v+m)
```

```python
cleaned_df = new_df
new_df.head()
```

```python
new_df = new_df.sort_values('weighted_average',ascending=False)
```

```python
new_df
```

```python
new_df =  new_df[['title','vote_count','vote_average','weighted_average','popularity']]
```

```python
new_df
```

```python
popularity = new_df.sort_values('popularity',ascending=False)
popularity.head(5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 
plt.figure(figsize=(12,6))
axis1=sns.barplot(x=new_df['weighted_average'].head(10), y=new_df['title'].head(10), data=new_df)
plt.xlim(4,10)
```

```python

plt.figure(figsize=(12,6))
ax=sns.barplot(x=popularity['popularity'].head(10), y=popularity['title'].head(10), data=popularity)

plt.title('Most Popular by Votes', weight='bold')
plt.xlabel('Score of Popularity', weight='bold')
plt.ylabel('Movie Title', weight='bold')
```

#### Recommendation based on Scaled Weighted Average and Popularity Score(50% to Both)

```python
from sklearn.preprocessing import MinMaxScaler
scaling = MinMaxScaler()
new_df = scaling.fit_transform(cleaned_df[['weighted_average','popularity']])
df = pd.DataFrame(new_df,columns=['weighted_average','popularity'])
df.head()
# new_df[0:10]
```

```python
cleaned_df[['normalized_weight_average','normalized_popularity']] = df
```

```python
cleaned_df.head()
```

```python
cleaned_df['score'] = cleaned_df['normalized_weight_average']*0.5 + cleaned_df['normalized_popularity']*0.5
sorted_df = cleaned_df.sort_values(['score'],ascending=False)
sorted_df[['title', 'score']].head()
```
